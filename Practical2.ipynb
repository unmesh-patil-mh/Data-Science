```python
sorted_score = sorted(df['Reading_Score'])
```

```python
sorted_score
```

```python
q1 = np.percentile(sorted_score, 25)
```

```python
q3 = np.percentile(sorted_score, 75)
```

```python
print(q1,q3)
```

```python
IQR = q3 - q1
```

```python
lwr_bound = q1 - (1.5*IQR)
```

```python
upper_bound = q3 + (1.5 *IQR)
```

```python
print(lwr_bound,upper_bound)
```

```python
r_outliers = []
for i in sorted_score:
    if(i<lwr_bound or i>upper_bound):
        r_outliers.append(i)
    print(r_outliers)
```
